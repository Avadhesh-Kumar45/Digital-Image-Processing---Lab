**Write a program to improve the contrast of an image using histogram
equalization. The prototype of the function is as below:
histogram_equalisation(input_Image, no_of_bins); The function
should return the enhanced image. Consider two low-contrast input
images. Study the nature of the output image quality in each case
by varying the number of bins.**

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image1 = cv2.imread("/content/Lenna.png")
image2 = cv2.imread("/content/pexels-pixabay-45201.jpg")

gray_image1 = cv2.cvtColor(image1, cv2.COLOR_BGR2GRAY)
gray_image2 = cv2.cvtColor(image2, cv2.COLOR_BGR2GRAY)

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(gray_image1, cmap='gray')
plt.title("Low-Contrast Image 1")
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(gray_image2, cmap='gray')
plt.title("Low-Contrast Image 2")
plt.axis('off')

plt.show()
```

**Applying Histogram Equalization Function**

```python
def histogram_equalisation(input_image, no_of_bins=256):
    hist, bin_edges = np.histogram(input_image.flatten(), bins=no_of_bins, range=[0, 256])
    cdf = hist.cumsum()
    cdf_m = np.ma.masked_equal(cdf, 0)

    cdf_m = (cdf_m - cdf_m.min()) * 255 / (cdf_m.max() - cdf_m.min())
    cdf = np.ma.filled(cdf_m, 0).astype('uint8')

    scaled_image = ((input_image / 255) * (no_of_bins - 1)).astype(np.uint8)
    enhanced_image = cdf[scaled_image]

    return enhanced_image
```

**Apply histogram equalization with different bin values**

```python
equalized_image1_256 = histogram_equalisation(gray_image1, no_of_bins=256)
equalized_image2_256 = histogram_equalisation(gray_image2, no_of_bins=256)

equalized_image1_64 = histogram_equalisation(gray_image1, no_of_bins=64)
equalized_image2_64 = histogram_equalisation(gray_image2, no_of_bins=64)

equalized_image1_32 = histogram_equalisation(gray_image1, no_of_bins=32)
equalized_image2_32 = histogram_equalisation(gray_image2, no_of_bins=32)
```

**Displaying Histogram Equalized images with  256 Bins**

```python
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(equalized_image1_256, cmap='gray')
plt.title("Histogram Equalized Image 1 (256 Bins)")
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(equalized_image2_256, cmap='gray')
plt.title("Histogram Equalized Image 2 (256 Bins)")
plt.axis('off')

plt.show()

```
