**1. Develop programs for following spatial filtering operations on a gray-scale image.
a. Averaging: Implement averaging filtering operations for different window sizes and study their effect on the output image quality. Write your observations on output image quality.
b. Weighted averaging: Implement weighted averaging filtering operations for different window sizes and study their effect on the output image quality. Write your observations on output image quality.
c. Median filtering: Implement weighted averaging filtering operations for different window sizes and study their effect on the output image quality. Write your observations on output image quality.
d. Max filtering
e. Min filtering**

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('/content/Lenna.png')
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(gray_image, cmap='gray')
plt.show()
```

**a. Averaging**

```python
def apply_averaging_filter(image, kernel_size):
    return cv2.blur(image, (kernel_size, kernel_size))

output1 = apply_averaging_filter(gray_image, 3)
output2 = apply_averaging_filter(gray_image, 5)
output3 = apply_averaging_filter(gray_image, 7)

plt.figure(figsize=(12, 6))
plt.subplot(1, 4, 1), plt.imshow(gray_image, cmap='gray'), plt.title("Original")
plt.subplot(1, 4, 2), plt.imshow(output1, cmap='gray'), plt.title("3x3")
plt.subplot(1, 4, 3), plt.imshow(output2, cmap='gray'), plt.title("5x5")
plt.subplot(1, 4, 4), plt.imshow(output3, cmap='gray'), plt.title("7x7")
plt.show()
```

**b. Weighted averaging**

```python
def apply_gaussian_filter(image, kernel_size):
    return cv2.GaussianBlur(image, (kernel_size, kernel_size), 0)

output1 = apply_gaussian_filter(gray_image, 3)
output2 = apply_gaussian_filter(gray_image, 5)
output3 = apply_gaussian_filter(gray_image, 7)

plt.figure(figsize=(12, 6))
plt.subplot(1, 4, 1), plt.imshow(gray_image, cmap='gray'), plt.title("Original")
plt.subplot(1, 4, 2), plt.imshow(output1, cmap='gray'), plt.title("3x3")
plt.subplot(1, 4, 3), plt.imshow(output2, cmap='gray'), plt.title("5x5")
plt.subplot(1, 4, 4), plt.imshow(output3, cmap='gray'), plt.title("7x7")
plt.show()
```

**c. Median filtering:**

```python
def apply_median_filter(image, kernel_size):
    return cv2.medianBlur(image, kernel_size)

output1 = apply_median_filter(gray_image, 3)
output2 = apply_median_filter(gray_image, 5)
output3 = apply_median_filter(gray_image, 7)

plt.figure(figsize=(12, 6))
plt.subplot(1, 4, 1), plt.imshow(gray_image, cmap='gray'), plt.title("Original")
plt.subplot(1, 4, 2), plt.imshow(output1, cmap='gray'), plt.title("3x3")
plt.subplot(1, 4, 3), plt.imshow(output2, cmap='gray'), plt.title("5x5")
plt.subplot(1, 4, 4), plt.imshow(output3, cmap='gray'), plt.title("7x7")
plt.show()
```

**d. Max filtering**

```python
def apply_max_filter(image, kernel_size):
    return cv2.dilate(image, np.ones((kernel_size, kernel_size), np.uint8))

output1 = apply_max_filter(gray_image, 3)
output2 = apply_max_filter(gray_image, 5)
output3 = apply_max_filter(gray_image, 7)

plt.figure(figsize=(12, 6))
plt.subplot(1, 4, 1), plt.imshow(gray_image, cmap='gray'), plt.title("Original")
plt.subplot(1, 4, 2), plt.imshow(output1, cmap='gray'), plt.title("3x3")
plt.subplot(1, 4, 3), plt.imshow(output2, cmap='gray'), plt.title("5x5")
plt.subplot(1, 4, 4), plt.imshow(output3, cmap='gray'), plt.title("7x7")
plt.show()
```

**e. Min filtering**

```python
def apply_min_filter(image, kernel_size):
    return cv2.erode(image, np.ones((kernel_size, kernel_size), np.uint8))

output1 = apply_min_filter(gray_image, 3)
output2 = apply_min_filter(gray_image, 5)
output3 = apply_min_filter(gray_image, 7)

plt.figure(figsize=(12, 6))
plt.subplot(1, 4, 1), plt.imshow(gray_image, cmap='gray'), plt.title("Original")
plt.subplot(1, 4, 2), plt.imshow(output1, cmap='gray'), plt.title("3x3")
plt.subplot(1, 4, 3), plt.imshow(output2, cmap='gray'), plt.title("5x5")
plt.subplot(1, 4, 4), plt.imshow(output3, cmap='gray'), plt.title("7x7")
plt.show()
```
