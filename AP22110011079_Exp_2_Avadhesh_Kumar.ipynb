**1) Create black (intensity=0) and white (Intensity=127) images (A) of size 1024x1024 pixels
that consist of alternative horizontal lines of black and white. Each horizontal line should
be 128 pixels in width. Create black and white images (B) of size 1024x1024 that consist of
alternative vertical lines of black and white. Each vertical line should be of 128 pixels in width.
Perform the following operations on Image A and Image B. Display the resulting image of
each operation.
a. Image addition of A and B (pixel-wise addition)
b. Subtraction of A and B (Pixel wise subtraction)
c. Multiplying Images of A and B (Pixel-wise multiplication)**

```python
import numpy as np
import matplotlib.pyplot as plt
```

 Create Image A (horizontal lines)



```python
image_A = np.zeros((1024, 1024), dtype=np.uint8)
image_A[np.arange(0, 1024, 256)[:, None] + np.arange(128), :] = 127  # Create horizontal stripes
```

Create Image B (vertical lines)

```python
image_B = np.zeros((1024, 1024), dtype=np.uint8)
image_B[:, np.arange(0, 1024, 256)[:, None] + np.arange(128)] = 127  # Create vertical stripes
```

Display Image A and B

```python
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.title("Image A (Horizontal Lines)")
plt.imshow(image_A, cmap='gray')

plt.subplot(1, 2, 2)
plt.title("Image B (Vertical Lines)")
plt.imshow(image_B, cmap='gray')
plt.show()
```

**Perform operations**

a. Image addition

```python
image_add = np.clip(image_A.astype(np.int16) + image_B.astype(np.int16), 0, 255).astype(np.uint8)
```

b. Image subtraction

```python
image_sub = np.clip(image_A.astype(np.int16) - image_B.astype(np.int16), 0, 255).astype(np.uint8)
```

c. Image multiplication

```python
image_mul = np.clip(image_A.astype(np.int16) * image_B.astype(np.int16), 0, 255).astype(np.uint8)
```

**Display results**

```python
plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
plt.title("Addition (A + B)")
plt.imshow(image_add, cmap='gray')

plt.subplot(1, 3, 2)
plt.title("Subtraction (A - B)")
plt.imshow(image_sub, cmap='gray')

plt.subplot(1, 3, 3)
plt.title("Multiplication (A * B)")
plt.imshow(image_mul, cmap='gray')
plt.show()
```

**2) Write a program to perform the following operations
a) Create a grayscale image of size 256x1024. The intensity of the image should vary sinusoidal
horizontally.
b) Create a white image of size 256x256, with the black box of size 58x58 at the centre.**

a) Grayscale image with sinusoidal intensity

```python
x = np.arange(1024)
intensity = 127 + 127 * np.sin(2 * np.pi * x / 1024)  # Sinusoidal intensity
image_sin = np.zeros((256, 1024), dtype=np.uint8)
for i in range(256):  # Repeat the intensity pattern for each row
    image_sin[i, :] = intensity
```

b) White image with black box at the center

```python
image_box = np.ones((256, 256), dtype=np.uint8) * 255  # White image
center_x, center_y = 128, 128
box_size = 58
start_x, start_y = center_x - box_size // 2, center_y - box_size // 2
image_box[start_x:start_x+box_size, start_y:start_y+box_size] = 0
```

```python
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.title("Grayscale Image (Sinusoidal Intensity)")
plt.imshow(image_sin, cmap='gray')

plt.subplot(1, 2, 2)
plt.title("White Image with Black Box")
plt.imshow(image_box, cmap='gray')
plt.show()
```
