# Perform the following operations using library functions
a. Read, Display and write any color image in other formats. b.
Find RED, GREEN and BLUE plane of the color image. c. Convert
color image to grayscale image and binary image
d. Resize the image by one half and one quarter. i.e. Image rotates
by 45, 90 and 180 degrees.

```python
import cv2
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
import numpy as np
```

```python
def disp_img(img, title="Image", cmap=None):
    plt.imshow(img, cmap=cmap)
    plt.title(title)
    plt.axis('off')
    plt.show()
```

```python
from google.colab import files
uploaded = files.upload()
```

**a. Read, Display and write any color image in other formats.**

```python
# Read the uploaded image
img_path = list(uploaded.keys())[0]
img = cv2.imread(img_path)
```

```python
# Display the uploaded image
disp_img(cv2.cvtColor(img, cv2.COLOR_BGR2RGB), "Original image")
```

```python
# Save the image in another format (e.g., PNG)
cv2.imwrite("out_img.png", img)
print("Image saved as out_img.png")
```

**b. Find RED, GREEN and BLUE plane of the color image.**

```python
B=img[:, :, 0]
G=img[:, :, 1]
R=img[:, :, 2]
```

```python
# Blue Plane
plt.imshow(np.dstack((np.zeros_like(B), np.zeros_like(B), B)))
plt.title("Blue Plane")
plt.axis('off')
plt.show()
```

```python
# Green Plane
plt.imshow(np.dstack((np.zeros_like(B), B, np.zeros_like(B))))
plt.title("Green Plane")
plt.axis("off")
```

```python
# Red Plane
plt.imshow(np.dstack((B, np.zeros_like(B), np.zeros_like(B))))
plt.title("Red Plane")
plt.axis("off")
```

**c. Convert
color image to grayscale image and binary image**

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
_, binary = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)
```

```python
disp_img(gray, "Grayscale Image", cmap="gray")
disp_img(binary, "Binary Image", cmap="gray")
```

**d. Resize the image by one half and one quarter. i.e. Image rotates
by 45, 90 and 180 degrees.**

```python
half_img = cv2.resize(img, (img.shape[1] // 2, img.shape[0] // 2))
quarter_img = cv2.resize(img, (img.shape[1] // 4, img.shape[0] // 4))
```

```python
print("Half-Size Image:")
cv2_imshow(half_img)
```

```python
print("Quarter-Size Image:")
cv2_imshow(quarter_img)
```

```python
(h, w) = img.shape[:2]
center = (w // 2, h // 2)
```

```python
rot_45 = cv2.getRotationMatrix2D(center, 45, 1.0)
rot_img_45 = cv2.warpAffine(img, rot_45, (w, h))
disp_img(cv2.cvtColor(rot_img_45, cv2.COLOR_BGR2RGB), "Rotated by 45 Degrees")
```

```python
rot_img_90 = cv2.rotate(img, cv2.ROTATE_90_CLOCKWISE)
disp_img(cv2.cvtColor(rot_img_90, cv2.COLOR_BGR2RGB), "Rotated by 90 Degrees")
```

```python
rot_img_180 = cv2.rotate(img, cv2.ROTATE_180)
disp_img(cv2.cvtColor(rot_img_180, cv2.COLOR_BGR2RGB), "Rotated by 180 Degrees")
```
