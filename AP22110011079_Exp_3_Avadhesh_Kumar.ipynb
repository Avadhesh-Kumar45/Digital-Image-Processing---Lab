**Develop programs for following intensity transformation operation
on a grayscale image. Collect any grayscale image from any
source. Process that image using these operations.
a. Image negative
b. Log transformation and inverse log transform: s = c log (1+r), c
is a const, r â‰¥ 0. s is the pixel intensity of the output image, and r is the pixel
intensity of the input image. Study the effect of constant c on the
quality of the output image.
c. Power law transformation: Study the effect of different values of
Gamma is used in this transformation.
d. Contrast stretching
e. Grey level slicing**

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('/content/gray.jpg', cv2.IMREAD_GRAYSCALE)

plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.show()
```

**a. Image Negative**

```python
image_negative = 255 - image

plt.imshow(image_negative, cmap='gray')
plt.title('Image Negative')
plt.show()
```

**b. Log Transformation**

```python
c = 255 / np.log(1 + np.max(image))
log_transformed = c * np.log(1 + image)
log_transformed = np.uint8(log_transformed)

plt.imshow(log_transformed, cmap='gray')
plt.title('Log Transformed Image')
plt.show()

```

**Inverse Log Transformation**

```python
inverse_log_transformed = np.exp(log_transformed / c) - 1
inverse_log_transformed = np.uint8(inverse_log_transformed)

plt.imshow(inverse_log_transformed, cmap='gray')
plt.title('Inverse Log Transformed Image')
plt.show()
```

**c. Power Law Transformation**

```python
gamma_values = [0.1, 0.5, 1.0, 2.0]

for gamma in gamma_values:
    power_law_transformed = np.power(image / 255.0, gamma) * 255.0
    power_law_transformed = np.uint8(power_law_transformed)

    plt.imshow(power_law_transformed, cmap='gray')
    plt.title(f'Power Law Transformation (Gamma = {gamma})')
    plt.show()
```

**d. Contrast Stretching**

```python
def contrast_stretching(img):
    min_val = np.min(img)
    max_val = np.max(img)
    stretched = (img - min_val) * (255.0 / (max_val - min_val))
    return np.uint8(stretched)

contrast_stretched = contrast_stretching(image)

plt.imshow(contrast_stretched, cmap='gray')
plt.title('Contrast Stretched Image')
plt.show()
```

**e. Grey Level Slicing**

```python
def grey_level_slicing(img, lower_bound, upper_bound):
    mask = (img >= lower_bound) & (img <= upper_bound)
    sliced = np.zeros_like(img)
    sliced[mask] = 255
    return sliced

lower_bound = 100
upper_bound = 200
grey_sliced = grey_level_slicing(image, lower_bound, upper_bound)

plt.imshow(grey_sliced, cmap='gray')
plt.title('Grey Level Sliced Image')
plt.show()
```
